    - sqft_living15 - Кв. метры жилой площади у 15 соседей

```python
df = pd.read_csv('./kc-house-data.csv', encoding='latin-1', sep=',')
df.head()
```

```python
df.info()
```

1. Создать новый признак price_per_sq_lot, который будет содержать стоимость за один кв. метр общей площади

```python
df['price_per_sq_lot']=df['price']/df['sqft_lot']
df[['price','sqft_lot','price_per_sq_lot']].sort_values('sqft_lot').head(10)
```

2. Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома.   
  Если реновации дома не было, то в новом признаке поставьте 0 

```python
df['delta_renovated'] = 0
df.loc[df['yr_renovated'] != 0, 'delta_renovated'] = df['yr_renovated'] - df['yr_built']
df.delta_renovated.sort_values()
```

3. Создайте признаки года продажи, месяца продажи 

```python
df['date'] = pd.to_datetime(df['date'])
df['year_sale'] = df['date'].apply(lambda x: x.date().year)
df['month_sale'] = df['date'].apply(lambda x: x.date().month)
df.sample(10)
```

4. Удалите признаки date, zipcode, lat, long

```python
df.drop(columns=['date','zipcode','lat','long'], inplace=True)
df.head()
```

#### Задание 3.
#####  Составьте несколько сводных таблиц

1.  Найдите среднюю стоимость домов в зависимости от количества спален. Отсортируйте от меньшей стоимости к большей


```python
df.groupby('bedrooms').agg({'price': 'mean'}).sort_values('price').apply(lambda x: round(x/ 1000,0))
```

2. Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома

```python
group_data = df.groupby('condition').agg({'price': ['min', 'mean', max]}).head()
```

```python
cols_list = group_data.columns.to_list()
group_data.sort_values(cols_list[2])
```

3. Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную и оценкой вида

```python
df.pivot_table(index='waterfront', columns='view', values='condition', aggfunc='count', fill_value=0)
```

4. Каких домов в зависимости от этажности и количества спален больше?

```python


df.pivot_table(index='floors', columns='bedrooms', values='id', aggfunc='count', fill_value=0)
```

5. Постройте таблицу с подсчетом медианной стоимости домов в данных в зависимости от состояния дома и оценки дома
