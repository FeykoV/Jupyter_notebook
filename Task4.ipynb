    #### Описание датасета


    - id - Уникальный ID для каждого дома
    - date - Дата продажи дома
    - price - Стоимость продажи дома
    - bedrooms - Кол-во спален
    - bathrooms - Кол-во ванных комнат (0.5 - туалет без душа)
    - sqft_living - Кв. футы жилые
    - sqft_lot - Кв. футы общие
    - floors - Кол-во этажей
    - waterfront - Есть набержная или нет
    - view - Значение от 0 до 4 насколько хороший вид
    - condition - Значение от 1 до 5 насколько хорошее состояние
    - grade - Значение от 1 до 13, где 1-3 плохая конструкция здания и дизайн, 7 - средний уровень конструкции и дизайна, 11-13 - высокое качество конструкции и дизайна
    - sqft_above - Кв. футы дома, которые находятся выше земли
    - sqft_basement - Кв. футы дома, которые находятся ниже земли
    - yr_built - Год постройки дома
    - yr_renovated - Год ремонта дома
    - zipcode - Индекс
    - lat - Широтма
    - long - Долгота
    - sqft_living15 - Кв. футы жилой площади у 15 соседей
    - sqft_lot15 - Кв. футы общей площади у 15 соседей

- Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
- Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
sns.jointplot(data=df, x = 'sqft_living15', y = 'price', kind='reg', height=12, ratio = 4);
```

```python
sns.jointplot(x=df['price'], y = df['grade'], kind = 'reg', height = 12, ratio = 4);
```

```python
g = sns.JointGrid(data=df, x='sqft_living', y = 'price', height = 10)
g.plot_joint(sns.scatterplot, s=100, alpha = 0.5)
g.plot_marginals(sns.histplot, kde = True);
```

```python
data = df.groupby('yr_built')['price'].mean().reset_index()
data
```

```python
plt.figure(figsize = (12, 6))
plt.scatter(data['yr_built'], data['price'])
plt.title('Распределение средних цен в зависимости от года постройки');
```

Вывод:
- распределение цены в зависимости от года постройки имеет U-образную форму;
- самые недорогие дома - дома 40х годов постройки;
- цена домов 80х годов постройки и выше начинает опять расти;
- скорей всего дома построенные больше 100 лет назад имеют историческую ценность, поэтому их стоимость, на одном уровне с ценой домов постройки в 21м веке.

```python
plt.figure(figsize =(10, 8))
sns.boxplot(x=df['grade'], y = df['price'], whis = 1.0)
plt.xlabel('Класс')
plt.ylabel('Цена')
plt.title('Распределение цен в зависимости от класса дома');
```

Вывод: 
- с увеличением уровня класса дома растет разбег по его цене.
- для домов с 7-11 классом наблюдается максимальное количество выбросов.

```python
plt.figure(figsize = (14,10))
sns.boxplot(x = df['price'], y = df['waterfront'].astype('str'), whis = 1.5, hue = df['grade'])
plt.title('Распределение цен по береговой линии')
plt.xlabel('Цена')
plt.ylabel('Наличие набережной');
```

Вывод:
- уровень цен домов не выходящих на набережную значительно ниже цены домов имеющих выход на нее.
- разбег цены домов с набережной больше, чем для домов без нее.

```python
plt.figure(figsize = (14,10))
sns.boxplot(x = df['price'], y = df['floors'].astype('str'), whis = 1.5)
plt.xlabel('Цена')
plt.ylabel('Этажность')
plt.title('Зависимость стоимости дома от его этажности')
```

Вывод:
- наибольший разбег по ценам домов наблюдается для 2,5-х этажей.
- наименьший разбег по ценам у домов в 3,5 этажа, так же стоимость этих домов практически не имеет выбросов.
- по 2-х этажным домам наблюдается максимальное количество выбросов.
- самый дорогой дом имеет 2,5 этажа и стоит почти 8000000 д.е.

```python
plt.figure(figsize = (14,10))
sns.boxplot(x=df['price'], y = df['condition'].astype('str'), whis = 1.5)
plt.xlabel('Цена')
plt.ylabel('Состояние дома')
plt.title('стоимость/Оценка состояния');
```

Вывод:
- дома, состояние которых оценивается в 1 и 2 почти не имеют выбросов.
- максимальное количество выбросов у домов с состоянием 3 и 4.
